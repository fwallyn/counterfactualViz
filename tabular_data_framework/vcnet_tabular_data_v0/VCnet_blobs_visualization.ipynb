```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Load data
save_name = "blobs" + "_test"
X = np.loadtxt("toy_dataset_plot/" + save_name)
C = np.loadtxt("toy_dataset_plot/counterfactuals_" + save_name )
y_x = np.loadtxt("toy_dataset_plot/predicted_examples_classes_" + save_name)
y_c = np.loadtxt("toy_dataset_plot/predicted_counterfactuals_classes_" + save_name)
y_true_x = np.loadtxt("toy_dataset_plot/true_examples_classes_" + save_name)

# Load meshgrid 
Z = np.loadtxt("toy_dataset_plot/contourf_" +  save_name) 
xx = np.loadtxt("toy_dataset_plot/xx_" +  save_name) 
yy = np.loadtxt("toy_dataset_plot/yy_" +  save_name) 
```

```python
plt.figure(figsize=(7,7))
c = plt.contourf(xx, yy, Z, cmap="RdBu")
plt.scatter(X[:,0],X[:,1],c=y_x)
plt.colorbar(c)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("Test data")
plt.show()
```

```python
plt.figure(figsize=(7,7))
c = plt.contourf(xx, yy, Z, cmap="RdBu")
plt.colorbar(c)
plt.scatter(C[:,0],C[:,1],c=y_c)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("Corresponding counterfactuals with new VCnet version")
plt.show()
```

```python
classe = 1
index = np.where(y_x==classe)
c = plt.contourf(xx, yy, Z, cmap="RdBu")
plt.colorbar(c)
plt.scatter(X[index][:, 0], X[index][:, 1], marker=".", s=10,label="Examples to explain")
plt.scatter(C[index][:,0],C[index][:,1],s=10,marker="x",label='Counterfactuals')
plt.title('Counterfactuals for example of class {}'.format(str(classe)))


subsample=np.random.choice(len(X[index]),20) #select 20 arrows to draw
plt.quiver(X[index][subsample, 0], X[index][subsample, 1], 
           C[index][subsample, 0]-X[index][subsample, 0], C[index][subsample, 1]-X[index][subsample, 1],
           angles='xy', #the way to compute the direction of the arrow
           scale=1,     #full length of the arrow
           width=0.003  #thickness of the arrow line
          )



plt.xlim([0, 1])
plt.ylim([0, 1])
plt.legend()
plt.show()
```
